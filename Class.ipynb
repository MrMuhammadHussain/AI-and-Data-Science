Combine Array

```python
a = np.arange(20).reshape(5,4)
print(a)
a.shape
```

```python
b = np.arange(20).reshape(5,4)
print(b)
b.shape
```

```python
np.concatenate((a,b))#Row Add Vartical Scale
```

```python
np.concatenate((a,b),axis=1) #coulum Add Horizontal Scale
```

```python
np.stack((a,b)) # also Commbine Arrys but dimantion Change both Arrys in 2d, and ans in 3d Array
```

```python
np.hstack((a,b))# onlny Commibe aarys like *concatenate*
```

```python
np.vstack((a,b)) # onlny Commibe aarys like *concatenate*
```

```python
combineArryConcat = np.concatenate((a,b))
combineArryStack = np.stack((a,b))
print(combineArryConcat ,"\n C<  >S \n",combineArryStack)
```
